```python
#creating loyalty tag for customers
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(max)
```

```python
df_ords_prods_merge.head(15)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

# 6. targeting spenders 

```python
# creating a spending flag based on average price across all orders
df_ords_prods_merge.groupby('user_id').agg({'prices': ['mean', 'min', 'max']})
```

```python
#creating column for average spending level
df_ords_prods_merge['spending_level'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# testing new column
df_ords_prods_merge.head(15)
```

```python
# displaying totals for customers either high or low spender
df_ords_prods_merge.loc[df_ords_prods_merge['spending_level'] >= 10, 'loyalty_flag_spending'] = 'High_spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['spending_level'] < 10, 'loyalty_flag_spending'] = 'Low_spender'
```

```python
df_ords_prods_merge['loyalty_flag_spending'].value_counts(dropna = False)
```

6. Most customers are considered low spenders
   

# 7. frequency of customers

```python
# finding regularity of customers based on median
df_ords_prods_merge['days_since_prior_order'].median
```

median is 3.0

```python
df_ords_prods_merge.groupby('user_id').agg({'days_since_prior_order': ['mean']})
```

```python
#creating new column for median days since prior order
df_ords_prods_merge['median_days_since_prior'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
df_ords_prods_merge.head(100
                    )
```

```python
# creating categories for customers based on median days since prior order
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior'] > 20, 'loyalty_flag_freq'] = 'Non-frequent_customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_prior'] > 10) & (df_ords_prods_merge['median_days_since_prior'] <= 20), 'loyalty_flag_freq'] = 'Regular_customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior'] < 10, 'loyalty_flag_freq'] = 'Frequent_customer'
```

```python
# display results
df_ords_prods_merge['loyalty_flag_freq'].value_counts(dropna = False)
```

7. most customers are frequent customers, folowed by a smaller qty of regular and non-frequent customers
